```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris)
```

```python
sns.lineplot(x='sepal_width',y='petal_width',data=iris)
```

```python
sns.barplot(x='sepal_length',y='species',data=iris)
```

```python
sns.distplot(iris['sepal_length'])
```

```python
sns.boxplot(x='sepal_width',y='petal_length',data=iris)
```

```python
sns.catplot(x='petal_length',y='species',data=iris)
```

```python
sns.pairplot(iris)
```

```python
sns.heatmap(iris.corr(),annot= True)
```
